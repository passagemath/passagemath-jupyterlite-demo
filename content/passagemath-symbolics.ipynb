```python
x = var('x')
eqn = (x-1)**2 <= x**2 - 2*x + 3
eqn.subs(x == 5)
```

```python
var('x,y,z')
M = matrix(2,2,[x,y,z,x])
v = vector([x,y])
v*M
```
