```python
from sage.combinat.e_one_star import E1Star, Face, Patch
```

```python
x = [Face((0,0,0),1), Face((0,0,0),2), Face((0,0,0),3)]
P = Patch(x)
P
```

```python
P.plot()
```

```python
sigma = WordMorphism({1:[1,2], 2:[1,3], 3:[1]})
E = E1Star(sigma)
E(P)
```

```python
E(P).plot()
```

```python
P = Patch([Face([0,0], 1), Face([0,0], 2)])
E = E1Star(WordMorphism({1:[1,2],2:[1]}))
F = E1Star(WordMorphism({1:[1,1,2],2:[2,1]}))
E(P,5).plot()
```
